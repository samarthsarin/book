```python
from flask import Flask,render_template,request,Markup
import re
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import warnings
warnings.filterwarnings('ignore')


model = load_model('C:/Users/Samarth/Desktop/Precily/Final Models/model_bilstm_multilabel_17 Dec.h5')
tokenizer = Tokenizer()
with open('C:/Users/Samarth/Desktop/Precily/Final Models/tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
            
app = Flask(__name__)


def color(keywords,sentence):
    f = '<span style="background-color: #99BE1B">{0}</span>'
    
    formattedText = ''
    words = keywords
    for i in words:
        try:
            k = re.search(i,sentence)
            my_sentence = sentence[k.span()[0]:k.span()[1]]
            sentence = sentence.replace(my_sentence,f.format(my_sentence))
        except:
            pass
    return sentence

def model_predict(text,model,tokenizer):
    cols= ['persons covered treaty eligibility',
 'pe',
 'business profits',
 'shipping air transport',
 'interest',
 'royalty',
 'royaltyfts',
 'fts',
 'independent personal services',
 'dividend',
 'capital gains',
 'other treaty issues',
 'gaar',
 'income not includible in total income',
 'profits gains from business profession',
 'other sources',
 'unexplained income',
 'chapter via deductions',
 'minimum alternate tax',
 'prosecution',
 'interest (interest on income tax refund)',
 'carry forward of losses',
 'penalty',
 'dispute resolution',
 'section 9',
 'refund processing section 143(1d)',
 'dividend distribution tax']
    #tokenizer = Tokenizer()
#     with open('C:/Users/Samarth/Desktop/Precily/Final Models/tokenizer.pickle', 'rb') as handle:
#         tokenizer = pickle.load(handle)
        
    b = tokenizer.texts_to_sequences([text])
    b_pad = pad_sequences(b,maxlen=500)
    results = model.predict(b_pad)
    answers = []
    for i in range(len(results[0])):
        if results[0][i]>0.7:
            answers.append(cols[i])

    return answers


@app.route('/',methods = ['GET','POST'])
def homepage():
        
    if request.method == 'POST':
        keywords = [
 'section 9\(1\)\(vi\)',
 'license',
 'equipment royalty',
 'process royalty',
 'copyright vs copyrighted article',
 'patent',
 'beneficial owner of property',
 'use or right to use ',
 'dry lease vs wet lease, ',
 'container leasing, ',
 'use of simulators, ',
 'forbearance fees, ',
 'use of natural resources',
 'copyright of a literary, artistic or scientific work',
 'trademarks',
 'secret formula or process',
 'design or model',
 'industrial, commercial or scientific experience',
 'provision of know how vs provision of services',
 'payment for leased line circuit/ connectivity charges',
 'content delivery solutions',
 'payment for know-how, designs, engineering drawings etc.',
 'fees for use of satellite',
 'trade mark , invention, scientific knowledge, experience, broadcasting, cable, optic fibre',
 'software royalty',
 'copyright of literary, artistic or scientific work',
 "nature of transferee's right",
 'shrink wrap software',
 'partial transfer of rights in the underlying copyright ',
 'complete transfer of rights in the underlying copyright',
        'section 9',
 'secret formula',]
        
        
        text = request.form['text']
        text = text.lower()
        answer = model_predict(text,model,tokenizer)
        coloredtext = color(keywords,text)
        return render_template('main.html',coloredtext = Markup(coloredtext),answer = answer)
    
    return render_template('main.html')
```

```python
if __name__ =='__main__':
    app.run()
```
